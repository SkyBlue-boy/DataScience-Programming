```python
import pandas as pd
from datetime import datetime

# 예시 데이터 프레임을 생성합니다.

# 월 이름을 숫자로 변환하기 위한 딕셔너리를 만듭니다.
month_dict = {
    'Jan': '01',
    'Feb': '02',
    'Mar': '03',
    'Apr': '04',
    'May': '05',
    'Jun': '06',
    'Jul': '07',
    'Aug': '08',
    'Sep': '09',
    'Oct': '10',
    'Nov': '11',
    'Dec': '12'
}

# '일시' 열의 값들을 반복하여 처리하고, 새로운 '변환된_일시' 열을 만듭니다.
result['변환된_일시'] = result['일시'].apply(lambda x: '20' + x.split('.')[1] + '-' + month_dict[x.split('.')[0]])

# 결과를 출력합니다.
print(result)
```

```python
from sklearn import ensemble  

params={"n_estimators": 500, "max_depth": 4, "min_samples_split": 5, "learning_rate": 0.01,"loss": "squared_error"}
model = ensemble.GradientBoostingRegressor(**params)
```

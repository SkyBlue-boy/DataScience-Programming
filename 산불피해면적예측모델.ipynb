```python
py_reg = setup(data = dt, target = 'damagearea', session_id = 232)
```

```python
top5_models = compare_models(n_select=5, fold=5, sort = 'mae')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test , y_train, y_test = train_test_split(dt.iloc[:,:-1], dt.iloc[:,-1],
                                                     test_size=0.1,
                                                     shuffle=True,
                                                     random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn import ensemble

params={"n_estimators": 500, "max_depth": 4, "min_samples_split": 5, "learning_rate": 0.01,"loss": "squared_error"}
model = ensemble.GradientBoostingRegressor(**params)
```

```python
model.fit(X_train, y_train)
```

```python
model.predict([[30, 10, 9, 55]])  #강수량, 기온, 풍속, 습도

#3월 평균
# 산불 크게 나려면, 낮고, 크고, 크고, 낮고
```
